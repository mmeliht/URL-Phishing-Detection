```python
df = df[['Classifier / Method','Result ']]
```

```python
df = df.rename(columns={"Result ": "Result"})
```

```python
df.columns
```

```python
df['Classifier / Method'].value_counts()
```

```python
df["Classifier / Method"] = df["Classifier / Method"].str.replace(r',\s*$', '', regex=True)
```

```python
df['Classifier / Method'].value_counts()
```

```python
# 1 Random forest → Random Forest
df.loc[df["Classifier / Method"] == "Random forest", "Classifier / Method"] = "Random Forest"

# 2 Logistic regression → Logistic Regression
df.loc[df["Classifier / Method"] == "Logistic regression", "Classifier / Method"] = "Logistic Regression"

# 3 Naïve Bayes → Naive Bayes
df.loc[df["Classifier / Method"] == "Naïve Bayes", "Classifier / Method"] = "Naive Bayes"

# 4 Support Vector Machine → SVM
df.loc[df["Classifier / Method"] == "Support Vector Machine", "Classifier / Method"] = "SVM"

# 5 support vector classification (SVC) → SVM
df.loc[df["Classifier / Method"] == "support vector classification (SVC)", "Classifier / Method"] = "SVM"

# 6 Single class SVM → SVM
df.loc[df["Classifier / Method"] == "Single class SVM", "Classifier / Method"] = "SVM"

# 7 Gradient Boost → Gradient Boosting
df.loc[df["Classifier / Method"] == "Gradient Boost", "Classifier / Method"] = "Gradient Boosting"
```

```python
#satır sonu boşluklarını silmek için
df["Classifier / Method"] = df["Classifier / Method"].str.strip()
```

```python
# K-Nearest Neighbors → KNN
df.loc[df["Classifier / Method"] == "K-Nearest Neighbors", "Classifier / Method"] = "KNN"

# K-Nearest Neighbors (KNN) → KNN
df.loc[df["Classifier / Method"] == "K-Nearest Neighbors (KNN)", "Classifier / Method"] = "KNN"

# KNN (without entropy) → KNN
df.loc[df["Classifier / Method"] == "KNN (without entropy)", "Classifier / Method"] = "KNN"
```

```python
# Naive Bayes Kernel → Naive Bayes
df.loc[df["Classifier / Method"] == "Naive Bayes Kernel", "Classifier / Method"] = "Naive Bayes"
```

```python
# Multilayer Perceptron → MLP
df.loc[df["Classifier / Method"] == "Multilayer Perceptron", "Classifier / Method"] = "MLP"

# MLP neural network → MLP
df.loc[df["Classifier / Method"] == "MLP neural network", "Classifier / Method"] = "MLP"
```

```python
df['Classifier / Method'].value_counts()
```

```python
# DCNN → CNN
df.loc[df["Classifier / Method"] == "DCNN", "Classifier / Method"] = "CNN"
```

```python
# LGM → LightGBM
df.loc[df["Classifier / Method"] == "LGM", "Classifier / Method"] = "LightGBM"
```

```python
df['Classifier / Method'].value_counts()
```

```python
df.loc[
    df["Classifier / Method"].str.contains(
        r"DBN-SVM|CNN-LSTM|CNN\+LSTM|DTCNN-LSTM|GA\+CNN\+LSTM|PSO\+CNN\+LSTM|HHO\+CNN\+LSTM",
        na=False
    ),
    "Classifier / Method"
] = "Hybrid Deep Learning Models"
```
